<h1>Titanic Survival Analysis</h1>

<h4>Bar Plot(class):</h4>
<p>Shows the survival rates for each passenger class.
Observes that higher-class passengers had better survival rates.</p>
<h4>Bar Plot (Gender):</h4>
<p>Compares the survival rates of male and female passengers.
Highlights that females had a much higher survival rate.</p>
<h4>Histogram (Age Distribution):</h4>
<p>Shows the age distribution of passengers who survived and those who did not.
Includes a KDE overlay for smooth visualization.</p>
<h4>Correlation Heatmap:</h4>
<p>Analyzes the relationships between numeric features in the dataset.
Identifies which variables are most strongly correlated with survival.</p>
<h4>Bar Plot (Embarkation Town):</h4>
<p>Displays survival rates based on the town passengers embarked from.</p>

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.columns
```

```python
print(type(titanic))
```

```python
sns.set_context('notebook', font_scale=1.2)
sns.set_style('whitegrid')
```

```python
plt.figure(figsize=(4,4))
sns.barplot(x='class', y='survived', data=titanic)  
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(4, 4))
sns.barplot(x='sex', y='survived', data=titanic, palette='Greens',hue='sex')
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.xlabel('Gender')
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.histplot(data=titanic, x='age', hue='survived', kde=True, multiple='stack', palette='coolwarm', bins=30)
plt.title('Age Distribution by Survival Status')
plt.ylabel('Count')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(5, 5))
numeric_columns = titanic.select_dtypes(include=['float64', 'int64'])

correlation = numeric_columns.corr()
sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap of Titanic Dataset')
plt.show()
```

```python
plt.figure(figsize=(5, 3))

sns.barplot(x='embark_town', y='survived', data=titanic, palette='magma', hue="sex")
plt.title('Survival Rate by Embarkation Town')
plt.ylabel('Survival Rate')
plt.xlabel('Embarkation Town')
plt.show()
```
